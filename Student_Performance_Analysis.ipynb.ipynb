```python
df.isnull().sum()
```

```python
df.fillna("Unknown", inplace=True)
df.drop_duplicates(inplace=True)
df.columns = df.columns.str.replace(" ", "_").str.lower()
```

```python
df['average_score'] = (df['math_score'] + df['reading_score'] + df['writing_score']) / 3
```

```python
from IPython.display import display, HTML

display(HTML("""
<style>
.visual-header {
    font-family: 'Poppins', sans-serif;
    background: linear-gradient(270deg, #36d1dc, #5b86e5, #ff6a00);
    background-size: 600% 600%;
    animation: gradientMove 6s ease infinite;
    padding: 20px;
    border-radius: 15px;
    text-align: center;
    color: white;
    font-size: 28px;
    font-weight: bold;
    box-shadow: 0 10px 25px rgba(0,0,0,0.3);
}

/* Animation */
@keyframes gradientMove {
    0% {background-position: 0% 50%;}
    50% {background-position: 100% 50%;}
    100% {background-position: 0% 50%;}
}
</style>

<div class="visual-header">
📊 Visual Analysis Dashboard 🎨
</div>
"""))
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='gender', y='average_score', data=df)
plt.title("Average Score by Gender", fontsize=14)
plt.show()
```

```python
fig = px.histogram(df, x="average_score", color="gender",
                   title="🎯 Score Distribution",
                   nbins=20)
fig.show()
```

```python
fig = px.pie(df, names='test_preparation_course',
             title='📘 Test Preparation Distribution',
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
fig = px.box(df, x="gender", y="average_score",
             color="lunch",
             title="📊 Performance Analysis")
fig.show()
```
